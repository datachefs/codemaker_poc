```python
from IPython.html import widgets      # Widget definitions
from IPython.display import display   # Used to display widgets in the notebook
import json                           # Read json configuration files
import glob                           # Used by get_directory_files


class CodeMaker:

    def  __init__(self):
        self.create_display()

        
    def create_display(self):
        self.tab_json_config = self.read_tab_json(['load_data', 'explore_data', 'clean_data', 'analyze_data', 
                            'visualize_data', 'export_data'])
        self.set_input_type_info()
        self.codemaker_container = widgets.ContainerWidget(
                              children=[self.create_input_display(), self.create_output_display()])
        display(self.codemaker_container)
        
        # With IPython Notebook, you gotta add CSS classes after displaying the widgets, so we do all of the styling here
        self.style_codemaker()

        
    def read_tab_json(self, tab_filenames):
        '''Given a list of names of json files, pulls corresponding json to load tab names and tab information,
           and create a list of commands that are in them'''
        self.commands  = {}
        tab_configs = []
        for tab_config_file in tab_filenames:
            in_file = tab_config_file + '.json'
            with open(in_file) as json_file:
                json_config = json.load(json_file)
            tab_configs.append(json_config)
            navigation = json_config.get("tab_config").get("subnav")
            if navigation:
                 for command in navigation:
                    title = command.get("title")
                    self.commands[title] = command                   
            else:
                print "FYI,", in_file, "doesn't contain any commands"
        return tab_configs
        
      
        
# ----- Create the Input Display --------------------------------------------------------------------------------------------------
        
    def create_input_display(self):
        self.create_input_columns()
        self.create_tab_navigation()
        self.input_container = widgets.ContainerWidget(children=[self.tab_navigation])
        return self.input_container

    
    def create_input_columns(self):        
        #self.subnav = widgets.ContainerWidget(children=[widgets.HTMLWidget(value ="This is where</br>The Subnav</br>Goes")])
        self.gobutton = widgets.ButtonWidget(description = "go")
        self.gobutton.on_click(self.make_code)
        self.input_box = widgets.ContainerWidget(children=[
                       widgets.HTMLWidget(value = "Input Fields Go Here</br>Input Fields Also Go Here</br>")])  
        self.tutorial =  widgets.ContainerWidget(children=[widgets.HTMLWidget(value ="This is where</br>We put</b>The Tutorials")])

        
    def create_tab_navigation(self):
        self.tab_containers = []
        self.subnavs = []
        self.input_containers = []
        for tab_config in self.tab_json_config:
            #print "\n\ntab config is ", tab_config
            subnav = widgets.ContainerWidget(children = [self.create_commands(tab_config)])
            #print "children are", subnav.children[0]
            test = subnav.children[0]
            #print "children of that container are", test.children
            self.subnavs.append(subnav)
            input_box_container = widgets.ContainerWidget(children=[self.input_box, self.gobutton] )
            input_container = widgets.ContainerWidget(children=[subnav,input_box_container,self.tutorial] )
            self.input_containers.append(input_container)
            self.tab_containers.append(input_container)
        self.tab_navigation = widgets.TabWidget(children=self.tab_containers)
                

    def create_commands(self, tab_config):
        self.command_links = []
        navigation = tab_config.get("tab_config").get("subnav")
        if navigation:
            for command in navigation:
                b = widgets.ButtonWidget(description=command.get("title"))
                b.on_click(self.load_command)
                self.command_links.append(b)
        else:
            self.command_links.append(widgets.ButtonWidget(description="Meh! No Commands!"))
        return widgets.ContainerWidget(children = self.command_links)


    def load_command(self,clicked_button):
        """ When a command is clicked, load the command"""
        self.command_name = clicked_button.description
        command_config =  self.commands[self.command_name]
        self.update_tutorial(command_config.get("tutorial"))
        self.update_input_box(command_config)
       
    def update_tutorial(self, new_tutorial):
        if not new_tutorial:
            new_tutorial = ''
        self.tutorial.children[0].value = new_tutorial
         
        
# ----- Commands:  code for creating input fields and creating the code when the person hits the "go" button -----
   
    def update_input_box(self, command_config):
        if command_config.get('special_widget_type',''):
            self.update_with_special_widget(command_config)
        else:
            self.update_input_fields(command_config.get("input"))

    def update_with_special_widget(self, command_config):
        special_widget_type = command_config.get('special_widget_type')
        print "Special widget type is", special_widget_type
        if special_widget_type in self.create_container_inputs:
            print "special widget function is ", self.create_container_inputs[special_widget_type]
            input_method = self.create_container_inputs[special_widget_type]
            self.input_box.children = input_method()
        else:
            print "can't find the Special widget type"
            self.input_box.children = [ widgets.HTMLWidget(value = "hey everybody, it's a special widget!") ]            

    def update_input_fields(self, new_input_box_config):
        if new_input_box_config:
            fields = []
            for field in new_input_box_config:
                widget_type = field.get('widget_type','')
                if widget_type in self.create_inputs:
                    input_method = self.create_inputs[widget_type]
                    fields.append(input_method(field))
                else:
                    fields = fields.append(widgets.HTMLWidget(value = 
                            "{0} isn't a widget type I recognize<br/>\n".format(field.get('label',''))))
            self.input_box.children = fields
        else:
            self.input_box.children = [ widgets.HTMLWidget(value = "") ]

            

    def make_code(self, button):
        command_config =  self.commands[self.command_name]
        if command_config.get('special_widget_type',''):        
            self.code.value = self.make_code_with_special_widget(command_config)
        else:
            self.code.value = self.make_code_with_fields(command_config)

            
    def make_code_with_special_widget(self, command_config):
        special_widget_type = command_config.get('special_widget_type')
        print "\nSpecial widget type is", special_widget_type
        if special_widget_type in self.create_container_inputs:
            code_generation_method = self.create_special_widget_code[special_widget_type]
            input_values = self.get_widget_values(self.input_box.children)
            print "input values are:", input_values
            return  code_generation_method(input_values)
        else:
            print "can't find the Special widget type"
            return "can't find the Special widget type, so can't produce code"

    def get_widget_values(self, input_widgets):
        """Return a list of the values the user types into the widgets"""
        values = []
        for w in input_widgets:
            values.append(w.value)
        return values        

    
    def make_code_with_fields(self, command_config):        
        field_names = self.get_field_names(command_config)
        code_template = command_config.get('template','')
        print code_template
        if code_template:
            return code_template.format(**field_names)
        else:
            return "OOPS:  {0} is missing its code template".format(self.command_name)       
        # NOTE:  I need to catch:  KeyError: 'dataframe2'  ; if so, put in, hey you're missing a field called dataframe2

        
    def get_field_names(self, command_config):
        #return self.command_name
        field_names = {}
        #print command_config, " is config for ", self.command_name
        input_fields =  command_config.get('input','')
        input_widgets = self.input_box.children
        if input_fields:
            for i in range(len(input_fields)):
                #print input_fields[i]
                field_name = input_fields[i].get('field_name','')
                #print field_name
                field_names[field_name] = input_widgets[i].value
        return field_names


    
# ----- Commands:  methods to create fields for input and methods for special widgets, for both input and generating the code -----   
    
    def set_input_type_info(self):
        """ Create the dictionaries that connect the JSON to methods for building the input fields and creating the code"""
        self.create_inputs = {'text': self.create_text_input}
        self.create_container_inputs = {'input_file': self.create_read_file_widgets}
        self.create_special_widget_code = {'input_file': self.create_read_file_code}
    
    def create_text_input(self, field):
        return widgets.TextWidget(description = field.get('label',''))
    
    
    # With these widgets, they replace everything in the container (explain)
    
    def create_read_file_widgets(self):
        """ Create the widgets needed to read in a file"""
        dataframe = widgets.TextWidget(description = "dataframe")
        print 'files in the directory', self.get_directory_files() 
        input_file_dropdown =widgets.DropdownWidget(
            values=self.get_directory_files(),
            description="Pick a file",
        )
        input_file = widgets.TextWidget(description = "Or enter name")
        return [dataframe,input_file_dropdown,input_file]
    
    
    def get_directory_files(self):
        """Get a list of all files in the current directory that are txt, csv, or xls/xlsx"""
        file_types=('*.txt','*.csv','*.xls','*.xlsx')
        files_grabbed=[]
        for file_type in file_types:
            files_grabbed.extend(glob.glob(file_type)) #Extend is like append, only it takes a list and pulls out the individual items
        return files_grabbed

    def create_read_file_code(self, input_values):
        print "\n\nTime for reading file code!"
        [dataframe,input_file_dropdown,input_file] = input_values
        print dataframe,input_file_dropdown,input_file
        if input_file == "":
            input_file = input_file_dropdown
        file_format = input_file.rsplit('.')[-1]
        print "File format is ", file_format
        if file_format == "csv":
            return"{frame} = pd.read_csv('{file}')".format(frame=dataframe,file=input_file)
        elif file_format == "lsx" or file_format == "xls":
            return "{frame} = pd.read_excel('{file}', 0, index_col=None, na_values=['NA'])".format(frame=dataframe,file=input_file)
        elif file_format == "txt":
            return "{frame} = pd.read_table('{file}')".format(frame=dataframe,file=input_file)
        else:
            return "I don't know what type of file this is. I can only take the following: .txt, .csv, .xls or .xlsx. No .jpeg sorry"

            
            
# ----- Create the Output Display --------------------------------------------------------------------------------------------------
   
    def create_output_display(self):
        self.header = widgets.LatexWidget(value = "Here is your code!")
        self.code = widgets.TextWidget()
        self.footer = widgets.LatexWidget(value = "Now just copy the code and paste it in!")
        self.output_container = widgets.ContainerWidget(children=[self.header, self.code, self.footer])
        return self.output_container
        

        
# ----- Style the Display --------------------------------------------------------------------------------------------------

    def style_codemaker(self):
        self.codemaker_container.set_css({'border': '1px solid SlateGray', 'padding': '5px', 'width': '100%', 'height': '600px'}) 
        self.style_input_display()
        self.style_output_display()



    def style_input_display(self):
        """ Add titles to the tabs, add CSS classes & individual CSS to the input widges """
        self.input_container.set_css({ 'height':'65%', 'width':'100%'})
        self.tab_navigation.set_css({'width':'100%'})
        self.set_tab_navigation_title() 
        self.style_navigation_tabs()
        self.style_subnav()

    def set_tab_navigation_title(self):
        tab_position = 0
        for tab_config in self.tab_json_config:
            self.tab_navigation.set_title(tab_position,  tab_config.get('tab_config').get('title'))
            container = self.tab_navigation.children[tab_position].children
            container[1].set_css({'border-right':'1px solid grey', 'border-left':'1px solid gray', 'min-height':'150px'})
            tab_position = tab_position + 1
            
       
    def style_navigation_tabs(self):
        for navigation_tab in self.tab_containers:
            navigation_tab.remove_class('vbox')
            navigation_tab.add_class('hbox')
            navigation_tab.set_css({'border': '1px solid grey'}) 
        self.input_box.set_css({'padding': '5px', 'width': '550px'})
        self.tutorial.set_css({'width': '200px'})

    def style_subnav(self):
        """ Style each of the subnav containers, including making the buttons look like links"""
        for subnav_container in self.subnavs:
            subnav_container.set_css({'width': '270px'})
            subnav = subnav_container.children[0]    # Right now the subnav container only has one child
            for command in subnav.children:
                command.add_class("btn btn-link")
                command.set_css({'fontSize':"14px"})

                    
        
    def style_output_display(self):
        """ Add CSS classes & individual CSS to the output widges """
        
```
